# Filling the Missing Values — Imputation
 # (i) Numerical data - use mean

```python
df3 = df.copy()
```

```python
df3.head()
```

```python
mean_value = df3['reviews_per_month'].mean()
df3['reviews_per_month'] = df3['reviews_per_month'].fillna(mean_value)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

# Handling Categorical Data 

```python
df3.nunique()
```

```python
df3["last_review"].fillna(df3["last_review"].value_counts().index[0], inplace = True)
```

```python
df3.isna().sum()
```

```python
df4 = df.copy()
```

```python
df4.isna().sum()
```

```python
df4["last_review"].fillna("not_review", inplace = True)
```
